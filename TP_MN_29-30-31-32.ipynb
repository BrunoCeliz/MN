## 29)
Queremos los coeficientes $\omega_i$ y las abscisas $x_i$ para el método de cuadratura de Gauss de orden $n=2$ pero en el intervalo $x \in [0,1]$. <br>
Como nosotros sabemos qué hacer para el caso en el que el intervalo es $x \in [-1,1]$, transformemos desde nuestro intervalo original a este: <br>

### <center> $I = \int_{0}^{1}f(t) dt = \int_{-1}^{1}f(\frac{x+1}{2}) \frac{dx}{2}$ </center> <br>
Entonces, en vez de utilizar $x^k$ , utilizo $t^k = (\frac{x+1}{2})^k$ ; $k \in (0,1,2,3)$ para encontrar los coeficientes y abscisas. <br>
El sistema de 4 ecuaciones para nuestra 4 incógnitas es: <br>
### <center> $\int_{-1}^{1}\frac{dx}{2} = 1 = \omega_1 + \omega_2$ </center> <br>
### <center> $\int_{-1}^{1} \frac{x+1}{2} \frac{dx}{2} = \frac{1}{2} = \omega_1t_1 + \omega_2t_2$ </center> <br>
### <center> $\int_{-1}^{1} \frac{x^2+1}{4} + \frac{x}{2} \frac{dx}{2} = \frac{1}{3} = \omega_1t_1^2 + \omega_2t_2^2$ </center> <br>
### <center> $\int_{-1}^{1} \frac{x^3+3x^2+3x}{8} \frac{dx}{2} = \frac{1}{4} = \omega_1t_1^3 + \omega_2t_2^3$ </center> <br>
Despejando obtenemos que <br>
### <center> $\omega_1 = \omega_2 = \frac{1}{2}$ ; $t_1 = \frac{3-\sqrt{3}}{6}$ ; $t_2 = \frac{3+\sqrt{3}}{6}$ </center> <br>
Por lo que el método de cuadratura de Gauss de orden 2 para $x \in [0,1]$ queda:
<center> $I \approx \frac{1}{2}[f(\frac{3-\sqrt{3}}{6}) + f(\frac{3+\sqrt{3}}{6})]$ </center>

### <center> $I = \int_{0}^{1}e^x dx = \int_{-1}^{1} \frac{e^{(x+1)/2}}{2} dx$ </center> <br>

## 32)
Por último, obtengamos el valor numérico de la integral <br>
<center> $I = \int_{-1}^{1} \frac{dx}{x+2} = ln(3) \approx 1.0986$ </center> <br>
Utilizando los métodos del Rectángulo *(mid-point o promedio)*, Trapecio, Simpson, Simpson 3/8 y Cuadratura de Gauss de orden 2.

```python
x = np.linspace(-1,1,1000) #dom
y = 1/(x+2) #f(x)

plt.figure(figsize=(13,13))
plt.plot(x,y,color='k',lw=3,label='f(x)')
plt.fill_between(x,y,alpha=0.6,label='I = ln(3) ~ 1.0986')
plt.grid(True)
plt.xlabel('x',size=20)
plt.ylabel('f(x)',size=20)
plt.title('$(x+2)^{-1}$ entre $[-1,1]$',size=20)
plt.xticks(size=20)
plt.yticks(size=20)
plt.legend(fontsize=15,shadow=True);
```

```python
#Copiamos y pegamos lo hecho en "23-24-25"

def f(x): #La función
    y = 1/(x+2)
    return y

a = -1 #el Dominio
b = 1

def pts(n): #Los pts a interpolar para diferente orden
    X = np.empty(n)
    Y = np.empty(n)
    h = (b-a)/(n-1)
    
    for i in range(n):
        X[i] = a + i*h
        Y[i] = f(X[i])
        
    return X,Y,h

I_T = np.empty(5) #Haremos 5 órdenes

def p(x,X,Y,N): #Para definir los polinomios interpolatorios
    
    p = 0 #variable para la sumatoria
        
    for j in range(N+1): #sumatoria desde j hasta n-1
        
        l = 1 #variable para la productoria, adentro del loop de la sumatoria
        
        for i in range(N+1): #productoria para el l(x)
            
            if i == j: #en la productoria no entra el término i = j
                continue
            else:
                l = l*((x-X[i])/(X[j]-X[i])) #tengo que definir la lista de pts X y su f(x), Y.
                
            #print(l) revisamos cada término de l
                
        p += l*Y[j] #usando lista y el polinomio l_j, voy armando el polinomio p con n pts. 
        
    return p
```

No lo especificamos con anterioridad, asi que la Regla del Rectángulo es: <br>
<center> $I \approx (b-a)f(\frac{b+a}{2})$ </center> <br>
A los otros métodos ya los conocemos y definimos, así que proseguimos a calcular:

```python
#"Regla del rectángulo" (que no es la del trapecio) supongo que es la del mid-point:
# I ~ (b-a)*f((b+a)/2)

I = 2*f(0)
I_T[0] = I

print(f'El valor de la integral con el método del Rectángulo ("mid-point") es = {I}')

#Trapecio:

X,Y,h = pts(2)

I = (Y[0]+Y[1])*h/2
I_T[1] = I

print(f'El valor de la integral con el método del Trapecio es = {I}')

p1 = p(x,X,Y,1)

#Simpson:

X,Y,h = pts(3)

I = (Y[0]+4*Y[1]+Y[2])*h/3
I_T[2] = I

print(f'El valor de la integral con el método de Simpson es = {I}')

p2 = p(x,X,Y,2)

#Simpson 3/8:

X,Y,h = pts(4)

I = (Y[0]+3*Y[1]+3*Y[2]+Y[3])*h*(3/8)
I_T[3] = I

print(f'El valor de la integral con el método de Simpson 3/8 es = {I}')

p3 = p(x,X,Y,3)

#Cuadratura de Gauss de orden 2 (ya sabemos los coeficientes):

I = f(-1/np.sqrt(3)) + f(1/np.sqrt(3))
I_T[4] = I

print(f'El valor de la integral con el método de Cuadratura de Gauss es = {I}')

print(f'\nEl valor analítico de la integral es {np.log(3)}')
```
